# Get missing value

```python
missing_cols = []
for col in df.columns:
	num_missing_val = df[col].isnull().sum()
	percent_missing_val = (num_missing_val / len(df)) * 100
	print(f"Missing value of {col} = {percent_missing_val}%")
	if percent_missing_val > 0:
		missing_cols.append((col, percent_missing_val))
	
```
